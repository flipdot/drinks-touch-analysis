```python
%matplotlib inline
import pandas as pd
import pylab as plt
```

```python
df = pd.read_csv("drinks.csv", sep="\t")
df = df.iloc[:-1]
df['timestamp'] = pd.to_datetime(df['timestamp'])
len(df)
df.head()
```

```python
df[['timestamp','name']].groupby('name').count().plot.bar()
```
